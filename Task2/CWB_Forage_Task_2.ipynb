```python
import pandas as pd
import hashlib
import os

path = "/content/sample_data/mobile_customers.xlsx"
df = pd.read_excel(path)
df.head()
```

```python
# converting email as str
df['email'] = df['email'].astype(str)
print(type(df['email'].iloc[0]))  # This should print <class 'str'>

df['salary'] = df['salary'].astype(int)
print(type(df['salary'].iloc[0]))  # This should print <class 'int'>

```

```python
df.info()
```

## Removing unnecessary columns


```python
new_df = df.drop(columns=['name', 'Unnamed: 0', 'credit_card_number', 'credit_card_security_code', 'address', 'birthdate', 'credit_card_expire'], axis = 1)


def sub_x(val):
    l = len(val)
    sub = ""
    for i in range(l):
        sub += 'x'  # Constructing the new string by adding 'x' for each character
    return sub

# Mask identifiable information
new_df['email'] = new_df['email'].apply(lambda x: sub_x(x.split('@')[0]) + '@' + x.split('@')[1])
```

```python
# Adding a new column 'new_column' with default value 0
new_df['salary_range'] = ""
```

```python
# prompt: find the minimum and maximum value of salary

min_salary = new_df['salary'].min()
max_salary = new_df['salary'].max()

print("Minimum salary:", min_salary)
print("Maximum salary:", max_salary)
```

above 200000

```python
# converting salary range as str
# new_df['salary_range'] = new_df['salary_range'].astype(str)

def salary_range(salary):
    if salary >= 10000 and salary <= 50000:
        return '10,000 - 50,000'
    elif salary >= 50000 and salary <= 10000:
        return '50,000 - 10,000'
    elif salary >= 100000 and salary <= 200000:
        return '100,000 - 200,000'
    else:
        return 'above 200,000'



# adding salary range
new_df['salary_range'] = new_df['salary'].apply(salary_range)
```

```python
# remove salary column
new_df.drop('salary', axis=1, inplace=True)
new_df
```

```python
new_df.to_excel('output_df.xlsx', index=False)
```
