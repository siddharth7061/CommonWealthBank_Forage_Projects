```python
import pandas as pd
```

```python
# Importing Data set
path = "/content/sample_data/supermarket_transactions.xlsx"
df = pd.read_excel(path)
df.head()
```

## Across locations, how many apples were purchased in cash?

```python
df_cash = df[df["payment_method"] == "cash"]
df_cash_apples = df_cash[df_cash["product_name"] == "apple"]

print(df_cash_apples["quantity"].sum())
```

Total number of apples purchased in cash is  **117**

```python
df_cash_apples
```

```python
# Acquiring unique store location names and quantity, storing it in a custom dataframe

u_store = df_cash_apples["store"].unique()
custom_df = pd.DataFrame(columns=["store", "quantity", "amount"])
for i in u_store:
  custom_df.loc[len(custom_df)] = [i, df_cash_apples[df_cash_apples["store"] == i]["quantity"].sum(), 4.59*df_cash_apples[df_cash_apples["store"] == i]["quantity"].sum()]

custom_df
```

### Plotting the above data in a bar graph for better visuals

```python
import matplotlib.pyplot as plt

# Group the data by store location and count the number of apples purchased in cash
location_counts = df_cash_apples.groupby('store')['product_name'].count()

# Set the figure size (width, height in inches)
plt.figure(figsize=(25, 6))  # Increase the width and height as needed

# Create a bar graph
plt.bar(custom_df["store"], custom_df["quantity"])

# Set the title and labels
plt.title('Number of Apples Purchased in Cash by Location')
plt.xlabel('Location')
plt.ylabel('Number of Apples')

# Display the graph
plt.show()
```

## How much total cash was spent on these apples?

```python
apple_amount = "{:.2f}".format(custom_df["amount"].sum())
print(apple_amount)
```

Based on the analysis, the total cash spent on apples is  **537.03**

## Across all payment methods, how much money was spent at the Bakershire store location by non-member customers?

```python
df_bakershire = df[df["store"] == "Bakershire"]
df_bakershire_non_member = df_bakershire[df_bakershire["customer_type"] == "non-member"]

bakershire_amount = "{:.2f}".format(df_bakershire_non_member['total_amount'].sum())
print(bakershire_amount)
```

Based on the analysis, the total cash spent at Bakershire store location by non-member customers is  **2,857.51**
